# Proceso de Extraccion

```python
"""
    Librerias utilizadas
"""

import pandas as pd
import requests
import pymongo
from pymongo import MongoClient
import json
pd.set_option('display.max_rows', 10000000)
pd.set_option('display.max_columns', None)
```

```python
token='6622488331213745'
base_url= 'https://www.superheroapi.com/api.php/'+token
mongo_client = pymongo.MongoClient("mongodb://localhost:27017/")
super_db=mongo_client["super"]
super_col=super_db["supers"]

superheroes=[requests.get(f"{base_url}/{i}") for i in range(1,152)]

for supers in superheroes:
    print(supers.json())
    super_col.insert_one(supers.json())
```

# Proceso de Transformacion

```python
# Obtener los documentos de la colección.
documents = super_col.find()

# Conviertiendo los documentos en diccionarios
documents_list = [doc for doc in documents]

# Convertir el JSON a un DataFrame de Pandas
df = pd.json_normalize(documents_list)
```

```python
from neo4j import GraphDatabase
import json

# Conexión a la base de datos Neo4j
uri = "bolt://localhost:7687"
username = "your_username"
password = "your_password"

# Función para insertar los datos del JSON en Neo4j
def insert_data(tx, data):
    query = """
    CREATE (hero:Superhero {id: $id, name: $name, image_url: $image_url})
    CREATE (hero)-[:HAS_POWERSTATS]->(power:Powerstats $powerstats)
    CREATE (hero)-[:HAS_BIOGRAPHY]->(bio:Biography $biography)
    CREATE (hero)-[:HAS_APPEARANCE]->(appear:Appearance $appearance)
    CREATE (hero)-[:HAS_WORK]->(work:Work $work)
    CREATE (hero)-[:HAS_CONNECTIONS]->(conn:Connections $connections)
    """
    tx.run(query, **data)

# Carga el JSON
with open('ruta/del/archivo.json') as file:
    data = json.load(file)

# Establece la conexión con la base de datos
driver = GraphDatabase.driver(uri, auth=(username, password))
with driver.session() as session:
    session.write_transaction(insert_data, data)
```
